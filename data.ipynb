```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.io as pio
pio.renderers.default = "vscode"


#def formatar(valor):
    #return "R$ {:,.2f}".format(valor)

lista_arq = os.listdir('C:\VENDAS\database')
display(lista_arq)




                     
```

```python
total=pd.DataFrame()

for arquivo in lista_arq:

    if "Vendas" in arquivo:
        base=pd.read_csv(f'C:\VENDAS\database\{arquivo}')
        base=base.drop('Unnamed: 0',axis=1)
        base=base.drop('Unnamed: 8',axis=1)
        
        #print(arquivo)
        total = pd.concat([total, base], ignore_index = True)

total['Cliente'] = total['Primeiro Nome'] + [' ']+ total['Sobrenome']
total=total.drop('Primeiro Nome', axis=1)
total=total.drop('Sobrenome', axis=1)

#display(total.info())

print("*"*150)
print("*"*150)

display(total)

print("*"*150)
print("*"*150)

qtd_produto=total['Produto'].value_counts(normalize=True)
display(qtd_produto)




```

```python
## Preço Unitário e Qtd Vendida


produtos =  total.groupby('Produto').sum()
produtos = produtos[['Quantidade Vendida','Preco Unitario']].sort_values(by='Quantidade Vendida', ascending=False)


display(produtos)


```

```python
# Faturamento
import plotly.express as px # type: ignore



total['faturamento'] = total['Quantidade Vendida'] * total['Preco Unitario']
t_faturamento =  total.groupby('Produto').sum()
t_faturamento = t_faturamento[['faturamento']].sort_values(by='faturamento', ascending=False)
display(t_faturamento)


t_loja=total.groupby('Loja').sum()
t_loja = t_loja[['faturamento']]



print(total)

#display(t_loja)





```

```python
import plotly.express as px # type: ignore
import numpy as np # type: ignore
import matplotlib.pyplot as plt # type: ignore

print('-'*150)
print('LOJA QUE MAIS FATUROU')
print('-'*150)
print('*'*150)


t_loja=total.groupby('Loja').sum()
t_loja = t_loja[['faturamento']]
t_loja['faturamento'] = t_loja['faturamento'] #apply(formatar)
display(t_loja)
px.histogram(total, x='Loja',color='Loja', y='faturamento')
```

```python
#display(total)

print('-'*150)
print('PRODUTO QUE MAIS FATUROU')
print('-'*150)
print('*'*150)
t_faturamento['faturamento'] = t_faturamento['faturamento']
display(t_faturamento)

px.histogram(total, x='Produto',color='Produto', y='faturamento')
```

```python
df=total[["Produto","faturamento","Loja"]]
#print(df)

print('-'*150)
print('Tabela Indice: Produtos')
print('-'*150)
print('*'*150)

pivot_tableprod = df.pivot_table(
    index="Produto",
    columns="Loja",
    values="faturamento",
    aggfunc="sum"

)
print(pivot_tableprod)

print('-'*150)
print('Tabela Indice: Lojas')
print('-'*150)
print('*'*150)


pivot_tableloj = df.pivot_table(
    index="Loja",
    columns="Produto",
    values="faturamento",
    aggfunc="sum"

)
print(pivot_tableloj)


```
